## 03-2. while문(반복문)
- 참인 한 계속 반복/거짓이 되면 종료.

```python
#열 번 찍어 안 넘어가는 나무 없다는 속담을 만들어보자.
treehit=1
while treehit <=10:
    print('나무를 %d번 찍었습니다.'%treehit)
    if treehit == 10:
        print('나무 넘어갑니다.')
    treehit +=1
```

```python
prompt = '''
1. Add
2. Del
3. List
4. Quit
Enter number: ''' #전체가 문자열 
```

```python
number = 0
while number !=4: #number가 4만 아니면 계속 반복!
    print(prompt)
    number = int(input())
```

### while문 강제로 빠져나가기: break
- while문 안이 계속 참이면 계속 반복되는데, 참이더라도 만일 이런 경우(if ~ ~) 이면 빠져나오기(break)

```python
#커피 자판기: 자판기 내 커피가 충분히 있으면 동전을 넣었을 때 커피가 나온다.
#하지만 커피가 떨어졌다면 판매를 중단하고 '판매중지'라는 문구를 보여줘야 한다.
coffee=10 #True
coin = 1 #True
while coin:
    print('돈을 넣었으니 커피를 줍니다.')
    coffee -= 1
    print('남은 커피는 %d개입니다.'%coffee)
    if not coffee: #False 
        print('커피가 다 떨어졌으니 판매를 중지합니다.')
        break
```

```python
#실제 커피 자판기처럼 살 붙이기 (돈을 정확히 300원을 넣어야 커피를 만들어 줌.)
coffee=10 #True

while True:
    money = int(input('돈을 넣어 주세요.'))
    if money == 300:
        print('돈을 넣었으니 커피를 줍니다.')
        coffee -= 1
        print('남은 커피는 %d개입니다.'%coffee)
    elif money > 300:
        print('거스름돈은 %d원 입니다.'%(money-300))
        print('돈을 넣었으니 커피를 줍니다.')
        coffee -= 1
        print('남은 커피는 %d개입니다.'%coffee)
    else:
        print('다시 돈을 넣어 주세요.')
        print('남은 커피는 %d개입니다.'%coffee)
     
    if not coffee:
        print('커피가 다 떨어졌으니 판매를 중지합니다.')
        break
```

### 조건에 맞지 않는 경우 맨 처음으로 돌아가기: continue
- while문 안에 있는 게 참이면 계속 반복되고 거짓이 되면 빠져나가는데, 거짓일 때 완전히 빠져나가지 않고 맨 처음으로 돌아가는 것.

```python
#1~10까지의 숫자 중 홀수만 출력하기 
a=0
while a<10:
    a +=1
    if a%2==0: 
        continue #짝수면 맨 처음으로 돌아감 - 뒤의 print(a)가 수행되지 않음!
    print(a)
```

```python
#총 5명이 시험을 보았는데, 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 보여주시오.
marks = [90, 25, 67, 45, 80]
num=0

for m in marks:
    num +=1
    if m > 60:
        print('%d번은 합격입니다.'%num)
    else: print('%d번은 불합격입니다.'%num)
```

### continue: for문 안의 문장을 수행하는 도중 continue를 만나면 for문의 처음으로 돌아가기

```python
#앞의 예제를 이용해, 60점 이상인 사람에게는 축하 메세지를, 나머지 사람에게는 아무런 메세지도 전하지 않는 프로그램을 만들기
marks = [90, 25, 67, 45, 80]
num=0

for m in marks:
    num +=1
    if m>60:
        print('%d번은 합격입니다.'%num)
    else: continue 
```

```python
marks = [90, 25, 67, 45, 80]
num=0

for m in marks:
    num +=1
    if m<60: 
        continue #불합격인 사람은 출력하지 않고 넘어감-처음으로 돌아감.
    print('%d번은 합격입니다.'%num)
```

### range: 숫자 리스트를 자동으로 만들어 주는 함수로, for과 함께 많이 쓰임.

```python
#for과 range를 이용해 1부터 10까지 더하는 프로그램을 만들어보자.
sum1=0
for i in range(1,11):
    sum1 +=i
print(sum1)
```

```python
#앞서 만들어본 60점 이상 합격 코드 -> num 부분을 range를 이용해 간단하게 바꿔보자.
marks = [90, 25, 67, 45, 80]
for num in range(len(marks)):
    if marks[num] <60: continue
    print('%d번은 합격입니다.'%(num+1))
```

```python
#for, range를 이용해 2단~9단까지의 구구단을 출력해보자.
for i in range(2,10):
    for j in range(1,10):
        print(i*j, end=' ') #출력의 끝을 지정(여기선 2띄고4띄고6,...)
    print('') #단이 끝났을 때의 줄바꿈 
```

### 리스트 내포: 리스트 안에 for문 포함하기(유용)
- [표현식/ for ~ in ~/ if 조건]
- 여러개의 for, if를 쓰고 싶으면 [표현식 / for~ in~ / if 조건 / for~ in~ / if 조건 / ... ] 이런식으로 쓰고 엔터치고 써도 된다.

```python
#예시
a=[1,2,3,4]
result =[]
for i in a:
    result.append(i*3)
print(result)
```

```python
#위의 과정을 리스트 내포를 이용해 하면 훨씬 간단해짐.
[i*3 for i in a] #i*3 결과값이 자동으로 리스트안에 들어가 있음!!
```

```python
#만약 짝수에만 3을 곱해서 담고 싶다면 'if 조건'사용하기
[i*3 for i in a if i%2==0]
```

```python
#구구단의 모든 결과를 리스트 하나에 담아보자.
result = [i*j for i in range(2,10) for j in range(1,10)]
print(result)
```
